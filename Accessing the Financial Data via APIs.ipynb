# **Correlation**

Correlation can be throught of as a normalized version of covariance so that correlation measures both strenght and direction of the linear relationships between two random variables. Whereas covariance reflects the units of both random variables.

```
ρ = σ_xy/σxσy
```


If pxY >0, then X and Y are positively correlated

If PXY <0, then X and Y are negatively correlated

If PxY =0, then X and Y are non-linearly correlated (Maybe correlated in higher dimensions)

If PXY =1, then X and Y are perfectly positively correlated

If PXY =- 1, then X and Y are perfectly negatively correlated

````python
def SML(rf, ret, betas):
  asset_ret = [rf + (rm-rf) * i for i in betas]
  plt.plot(betas, asset_ret)
  plt.xlabel('Asset Beta')
  plt.ylabel('Asset Return')
  plt.title('Security Market Line')
  plt.plot(1, ret, 'ro')
````

````python
betas = [CAPM_result_AAPL.params[0], CAPM_result_INTC.params[0], CAPM_result_MSFT.params[0]]
rm = returns['GSPC'].mean()
risk_free_mean = risk_free.mean()
````

````python
SML(risk_free_mean, rm, betas)
````

Given the beta, if the stock return locates below the SML, this implies that stock has a higer return per unit of risk assumed. Hence, it can be conclused that the stock is overvalued and decide to sell the stock. Conversely, is the stock locates above the SML, stock has a lower return per unit of risk assumed, wich is considered to be undervalued and buying a stock would be wise choice.

````python
ret_exp = [0.001, 0.0025, 0.0009]
df_ret_exp = pd.DataFrame(ret_exp)
df_betas = pd.DataFrame(betas)
df_all = pd.concat([df_ret_exp, df_betas], axis=1)
df_all.columns = ['betas', 'ret_exp']
df_all.index = returns.columns[:3]
````

````python
df_all
````

````python
def over_valued(rf, rm, stock):
  x = rf + (rm-rf)*df_all['betas'].loc[stock]
  y = df_all['ret_exp'].loc[stock]

  if x < y:
    print('The stock is undervalued')
  elif x > y:
    print('The stock is overvalued')
  else:
    print('The stock is fairly priced')
````

````python
def sharpe(stocks, rf, n_assets):
  sharpe_ratios = {}
  for i,j in zip (stocks, range(n_assets)):
    sharpe_ratios[i] = round((returns.iloc[:, j].mean()- rf) /np.std(returns.iloc[:, j]), 2)
  return sharpe_ratios
````
